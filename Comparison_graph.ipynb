```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
accuracy_scores = [0.9471978392977718, 0.9328831870357867, 0.9282241728561783, 0.8908845374746792]
models = ['Logistic Regression', 'Random Forest', 'Naive Bayes', 'Decision Trees']
plt.bar(models, accuracy_scores)
plt.xlabel('Models')
plt.ylabel('Accuracy')
plt.title('Accuracy Comparison')
plt.ylim([0.8, 1.0])
plt.show()
```

```python
precision_scores = [0.9383303034540623, 0.9324749642346208, 0.9335624284077892, 0.9010908018867925]
recall_scores = [0.9516380655226209, 0.9258522727272728, 0.9158589689563141, 0.8661093794275999]
plt.plot(models, precision_scores, label='Precision')
plt.plot(models, recall_scores, label='Recall')
plt.xlabel('Models')
plt.ylabel('Score')
plt.title('Precision and Recall Comparison')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Metrics for each model
accuracy_scores = [0.9471978392977718, 0.9328831870357867, 0.9282241728561783, 0.8908845374746792]
precision_scores = [0.9383303034540623, 0.9324749642346208, 0.9335624284077892, 0.9010908018867925]
models = ['Logistic Regression', 'Random Forest', 'Naive Bayes', 'Decision Trees']

# Plotting the comparison
plt.figure(figsize=(10, 6))
plt.plot(models, accuracy_scores, marker='o', label='Accuracy')
plt.plot(models, precision_scores, marker='o', label='Precision')
plt.title('Accuracy vs Precision')
plt.xlabel('Models')
plt.ylabel('Scores')
plt.legend()
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```
